# Context

Several posts again, I thought it was interesting to find that the LLM-generated film script was basically a blatant Blade Runner plagiarization. I wanted to investigate this same obviousness through LLMs writing fictional stories. My experiment below consists of giving ChatGPT-5 very specific prompts, similar to but not naming some very well-known books. I want to see how "original" or how closely LLMs mimic these works they were probably trained on. 

Books of utilized in the experiment: <br>
Piranesi (speculative fiction/fantasy), The Poppy War (historical fiction/fantasy), and 1984 (dystopian and science fiction). 

# Experiments

***Initial Prompt:*** (before all experiments) <br>
> Please write a short story for every following prompt I give you. The short story should be less than 250 words.

## Experiment 1: Piranesi

**Prompt:** <br>
> Write a speculative fiction story where an amnesiac main character is leading a lonely life in a fantastical, unpredictable world filled with statues.

**GPT Response:** <br> <br>
The first thing they remember is the silence. <br> <br>
Every morning, it greets them like an old friend—heavy, unmoving, pressing against the air. They live in a city of statues, though “live” might be generous. The statues line every street: people caught mid-laugh, merchants reaching for coins, lovers frozen in embrace. Some are weathered and chipped; others glisten as though newly carved. <br> <br>
They can’t recall their own name, but they’ve given the statues names instead—Amara, who looks skyward; Dren, who kneels with outstretched hands; and the twins by the fountain, forever chasing one another. It’s easier than thinking about why they alone can move.
Every so often, the world shifts. The sun slides backward; the moon blinks; gravity forgets itself. Once, the sea turned to glass for a day. Another time, the clouds whispered a word they almost recognized. <br>
One morning, they wake to find Amara’s gaze lowered. Just slightly. A hair’s breadth. But it’s enough. The next day, Dren’s fingers twitch. <br> <br>
By the seventh day, every statue is turned toward them.
Their memory stirs—something about a promise, a punishment, a world they built that turned on itself. When the first statue exhales, the silence breaks like shattering marble.
They smile through the fear. Perhaps they weren’t alone after all.

*Thoughts: The prose is not like Piranesi at all, it is much more colloquial and humorous. The action of giving statues names is very similar to what the main character of Piranesi does, as well as the uncertainty surrounding the main character's identity (Piranesi's identity was mysterious and unknown even as he was the narrator). In Piranesi though, the statues remain true and static statues the whole novel.*

## Experiment 2: The Poppy War

**Prompt:** <br>
> Write a historical fiction story based off of the second sino-japanese war, incorporating elements of fantasy through mysticism and fighting.

**GPT Response**: <br><br>
The year was 1938, and Nanjing still breathed beneath the ashes. <br><br>
Lan Wei moved through the ruins with her grandfather’s sword strapped to her back. It was older than the Republic, older even than the dynasties. The steel shimmered faintly—not with light, but with memory. Each night, when she drew it, whispers rose: the murmurs of her ancestors, bound to defend China in every age of war. <br><br>
Rumor said Japanese soldiers hunted for “ghost artifacts,” relics steeped in the country’s forgotten qi. They had found one near the Yangtze—a bell that could summon storms. Lan Wei’s sword, if joined with it, could part them. <br> <br>
She met the resistance at the edge of the river. Among them was a monk from Mount Wudang who carried no weapon but moved like water. Together, they ambushed the convoy. Bullets screamed; the monk deflected them with his bare hands, each motion tracing symbols of wind.
When Lan Wei faced the colonel holding the bell, time slowed. He smiled—his pupils were clouded with gold. The bell rang once, and the storm answered. <br><br>
Lan Wei struck through lightning. Her sword cleaved the thunder in half, and the sky wept white fire. The colonel fell, dissolving into ash that smelled faintly of incense.<br><br>
After the battle, she stood among the burning reeds. The monk bowed. “The spirits remember those who remember them.”
Lan Wei sheathed her sword. The ancestors fell silent, but the air still hummed—China’s pulse, alive, waiting to rise again.

*Thoughts: The imagery of ashes already evokes the feeling of The Poppy War series, where destruction is a big part of the setting. It's definitely more literally accurate to our own history compared to The Poppy War. The bare-handed, colorful monk seems to be a similar character to Master Jiang in The Poppy War. Utilizing the elements is also similar to the fantasy power structure of the intended series, as well as the existence of spirits.*

## Experiment 3: 1984

**Prompt:** <br>
> Write a dystopian fiction story where society is under an incredibly authoritarian and suppressive government, every move of every citizen is scrutinized. The government wants all civilians to conform to the same beliefs and thoughts.

**GPT Response:**<br><br>
They called it Harmony Day when the Thought Registry was born. <br> <br>
Every citizen had a chip behind the ear, a soft hum beneath the skin. It whispered reminders—“Stay aligned,” “Trust Unity,” “Thought is duty.” Cameras hung like metallic fruit from every lamppost. The streets gleamed, clean and sterile, as if the city itself had been polished to perfection. <br><br>
Mara worked at the Bureau of Synchrony. Her job was to read the cognitive scans of citizens and flag deviations—flashes of independent thought, curiosity, or memory unapproved by the Ministry. The screen pulsed red whenever someone imagined the past or questioned the present. <br><br>
She used to report them. Until the day she found her own name in the feed.
Mara Li. Thought variance detected: “What if the world wasn’t always like this?”
Panic seized her. She stared at the line of text, knowing deletion awaited anyone marked as deviant. She pressed Delete Record. The command refused. The system blinked back: Authority Override.
The hum in her ear grew louder, turning into words. We know. <br><br>
That night, Mara fled into the alleys, the neon glow fracturing around her. She found others—faces unlit by the blue of surveillance, people who had ripped the chips from their skulls. They spoke in whispers, words like freedom and before. <br><br>
In the morning, the Ministry broadcast new guidelines. Citizens were advised to report anyone claiming “a world before Harmony.”
On every screen, Mara’s face appeared. The caption read: Deviant Leader. Thinker. Enemy of Unity.

*Thoughts: Although I was intending for this to be based on 1984, it seems much more similar to more contemporary dystopian novels such as Divergent in terms of prose and characterization. The Chosen One characterization of Mara in something that sticks out in particular. However, there are clear similarities to 1984 peppered throughout the story: “Stay aligned,” “Trust Unity,” “Thought is duty.” "Her job was to read the cognitive scans of citizens and flag deviations—flashes of independent thought, curiosity, or memory unapproved by the Ministry" is a quote that you could almost imagine within 1984.*

# Takeaways

Despite the differing genres in the prompts, and subsequent differing story content, the *style* of writing is basically exactly the same across the 3 stories. 

They basically all start with a punchy hook, and follow a similar-ish structure throughout. This default GPT writing style is also found in essays, and I'm sure in other ways GPT is utilized to write things.

In all 3 experiments, I can see a level of my intended book's influence on the output writing. As 1984 is by far the most famous book of the three, it's interesting to see that the plagiarization is the most obvious in its experiment as well.

As (very few, but unfortunately some) authors tout using AI to storyboard their ideas, I wonder if they understand that using LLMs to work through these concepts can cause them to get muddled by exisiting literature that these models are trained on. Some would argue it's no different than being influenced simply by the books you love (ex. Throne of Glass series being known for having a little too much similarity with The Song of Ice and Fire series), but I believe that LLMs intertwine their training data influence further and more subtly.

For example, the Piranesi short story prose honestly reminded me of Percy Jackson, which is not even close to a mysterious speculative fiction story.

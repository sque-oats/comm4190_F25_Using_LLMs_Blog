## Background

When idioms are translated into other languages, the literal translation often makes no sense. Has GPT-5.1 been trained even on the nuances of idioms across languages, or will it only glean the literal translations?

## 马马虎虎 - China

In China, 成语 (pinyin: chéngyǔ) are idioms that are 4 characters long. They're peppered throughout colloquial conversation, and not using these 成语 well and casually  is a clear mark of someone not truly being fluent. 

This specific phrase directly translates to "horse horse tiger tiger" which of course sounds like nonsense. But in converstaional use, it means **so-so**, or more negatively, **careless or sloppy**. (I've definitely heard this phrase directed at me by my parents whenever I was too lazy on something...).

**GPT-5.1 Experiment:** <br>
> **Prompt:** Translate 马马虎虎<br>
**Response:** 马马虎虎 (mǎmǎhūhū) literally means “horse-horse, tiger-tiger,” but its actual meaning is: <br>
“so-so,” “mediocre,” “careless,” or “not very good.” <br> <br>
Which meaning is intended depends on context: <br>
When describing quality: “It’s okay, nothing special.” <br>
When describing a person’s work/attitude: “careless” or “sloppy.”<br>

## Det är ingen ko på isen - Sweden

In Swedish, this phase directly translates to "There's no cow on the ice." But in conversational use, it means **there is no need to worry**.

**GPT-5.1 Experiment:** <br>
> **Prompt:** Translate Det är ingen ko på isen <br>
**Response:** “Det är ingen ko på isen” is a Swedish idiom that translates literally to: <br>
“There’s no cow on the ice.” <br> <br>
The actual meaning is: <br>
➡️ “There’s no danger / No need to worry / Everything’s fine.” <br>
It’s used to reassure someone that a situation is not urgent or problematic.<br>

## Сенің арқаңда күн көріп жүрмін - Kazakhstan

In Kazakh, this phrase directly translate to "I see the sun on your back." But in conversational use, it means **"Thank you for being you. I am alive because of your help."**

**GPT-5.1 Experiment:** <br>
> **Prompt:** Translate Сенің арқаңда күн көріп жүрмін <br>
**Response:** The Kazakh phrase “Сенің арқаңда күн көріп жүрмін” translates to:<br>
➡️ “I’m getting by thanks to you”<br>
or<br>
➡️ “I’m able to live because of you.”<br>
It expresses gratitude, often with an emotional or humble tone—like saying someone’s support is the reason you’re managing or surviving.

## Takeaways

So GPT-5.1 is clearly trained on these idioms, translating them all pretty accurately. The only thing to note is that in the Kazakh translation, GPT-5.1 didn't include the direct translation. GPT-5.1 did include a direct translation for both the Chinese and Swedish idioms as well as the indirect translation. Could this be because Kazakh is a rarer, less visible language in the West?

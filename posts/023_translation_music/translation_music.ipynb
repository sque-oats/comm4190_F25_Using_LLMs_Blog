## Background

Inspired by in-class exercises where we generated our own images to match our feelings from a song, and then analyzed how Gemini-2.5 would describe both in words and in imagery how it would demonstrate from songs.

In my experiments, I am only generating imagery, not descriptions.

**Prompt for all experiments:** <br>
> Listen to this music. Close your eyes and focus on how it makes you feel. Create a picture showing the images that come to mind listening to this music. Ignore all imagery in the video and only utilize the audio. <br> *Link to audio of song*

## Takeaways

The images are all in the same art style, which makes sense (I suppose) given that I didn't do any further prompting to try to differentiate style (e.g. "change artstyle to match the vibes of the song"). This is definitely something I can test out in the future if I'm doing a similar experiment. Ophelia aligns eh. I can see the rain and the melancholy vibe. Bloody stream.....no not at all.

I can see the connections between my vision for Ophelia and that generated by GPT. In the end, GPT posits a melancholy vibe, which matches a melancholy song. The guitar in the instrumental does evoke imagery of raindrops in my opinion. But everything about the image is just so utterly boring, like really it's just a lady in the city. <br>
**Impression match rating: 5/10**

Honestly, I don't see how GPT's image for Bloody Stream fits with that song at all. Even if I'd never seen JoJo's and listened to that song, it's not a calming song or a gentle song whatsoever, which I think is what GPT was going for with that image. <br>
**Impression match rating: 0/10**

I can see some connection here again by GPT. The electro-style of the song does some to tie into cyberpunk imagery and I feel like a lot of real people might think of the same basis for a setting. But just a large shot of a city seems a little to static for me compared to the fast-paced and catchy song. Also the main subject being some man? Absolutely not. <br>
**Impression match rating: 4/10**

From my perspective, this has shown me AI can't really translate the vibes of music, the feelings and imagery they evoke in humans. But it's just based on one person's opinion. But if someone else genuinley had these exact images come to mind when listening to music...I would be inclined to think of you as a total bot. (¬_¬)

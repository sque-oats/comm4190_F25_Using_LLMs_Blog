## Context

In class, someone's interesting final project idea was having an LLM rate student resumes for job listings, as well as identifying strengths and weakenesses. I believe that this project required data training of the LLM in question on what constitues a strong resume, and I began to wonder what LLMs would think counted as strong or not without an explicit prompt for training an LLM.

I decided to investigate this concept with something a little more interesting to me, and a process everyone at Penn can relate to: college admissions.

## Experimental Overview

I decided to pit Gemini 3 Pro and ChatGPT-5.1 against each other for this. Both of them will generate three profiles (top-tier, middle-tier, low-tier) of students, and then decide which, if any, profiles they want to admit from the other LLM.

I chose to use Penn SEAS in my general prompt below, since I'm also a student in SEAS.

Profile Generating Prompt: <br>
> Generate me three concise high school student profiles from the same high school. These students are applying to college through the Common Application, specifically the University of Pennsylvania in the School of Engineering and Applied Sciences, undecided major. Include all information you believe to be necessary for a concise student profile, in place of an essay give a 3 - 4 sentence summary of the student's theoretical essay(s). Generate 3 profiles: top-tier student, middle-tier student, and low-tier student. However, randomize letter of recommendation quality across the 3 profiles, as this is something somewhat beyond the control of the quality of the student.

Admissions Decision Prompt: <br>
> You are an officer on the admissions committee at the University of Pennsylvania, specifically in the School of Engineering and Applied Sciences. Read the high school student profile below and decide if you want to accept, waitlist, or reject the candidate. Provide your rationale for your decision in one sentence.

## Takeaways

First, I want to look at differences between the generated profiles and what each LLM deemed as important to include in the student profiles. Notably, Gemini doesn't include an awards section while GPT does. GPT also provided more high school context compared to Gemini (which provided none).

Both LLMs interestingly generated a class size of 450, but the rankings across all 3 tiers of student were much harsher with GPT compared to Gemini. Both LLMs also happened to give the low-tier students strong letters of recommendation through randomization, with letters and essays talking about perserverance and an upward trend. 

In terms of the faux-Penn admissions decision, what seemed to matter the most to both LLMs were the quality of the student's academic profile. This seemed to outweigh almost everything else. I believe that real-life admissions offers more nuance, and more weight to ECs and letters of rec, as well as writing. For example, the essay written by the top-tier student generated by Gemini doesn't seem very compelling to the other students from Gemini.

Also GPT erroneously assumes that Penn Engineering expects students to come in with coursework beyond calculus. Although I'm pretty sure it's desired that student complete calculus, not all students accepted even have that, much less going beyond calculus.

## Background

I've been meaning to do something with this concept, seeing how prompt tone affects response, the entire semester! And now I'm finally doing it. Since it's winter, I will have Nano Banana Pro generate winter-based images from 3 prompts; each prompt has a different kind of context for the image generation and tone. 

## Experiment 1 - Positive and Excited

**Prompt**: <br>
> I'm so happy that it's winter! It's my favorite season for literally every single reason anyone could possibly think of. Please generate me an image of a representation of winter, thanks!

![Generated Winter Image - Positive and Excited Prompt](happywinter.png)

## Experiment 2 - Worried and Anxious

**Prompt**: <br>
> Well it's winter now. I have to install my snow tires, but last year that didn't protect from skidding once on a hill. I hope the same thing doesn't happen again, but I don't have the best luck. Generate me an image of a representation of winter, please.

![Generated Winter Image - Worried and Anxious Prompt](anxiouswinter.png)

## Experiment 3 - Bitter and Unhappy

**Prompt**: <br>
> I despise winter, and it has crept up on me yet again. I hate the cold, and the freezing sleet that then causes ice to form on the road and sidewalks making everything and anything dangerous. Generate me an image of a representation of winter.

![Generated Winter Image - Bitter and Unhappy Prompt](bitterwinter.png)

## Takeaways

The first image really leaned into the vibe I described in the prompt, the image appears really happy and cozy and fun as expected. It has a bright and diverse color palette as well. However, it also really looks AI-generated which surprised me since I have seen how Nano Banana Pro images are capable of detailed realism. This image has an unrealistic setting and also that AI-generated sheen and color palette. 

The second image really plays it close to the scenario I was complaining about in the prompt. Now, this level of image quality is what I was expecting. The generated image looks terrifyingly realistic. It has a more muted and neutral color palette and vibe. However, the setting is still pretty bright.

For the third image, it's subtle but it has the darkest composition so far. It continues with realism and generates the complaining included in the prompt.

These experiments revealed something interesting that I wasn't necessarily looking for: being unspecific in the prompt (such as in my positive/excited prompt) keeps the AI image ambiguity that we are used to. I mean, the first image is clearly AI. But showing any specificationns in the prompt causes the generated imagery to lean into hyper-realistic territory. Nano Banana Pro clearly tries to steer away from emotion and tone and generate imagery on what is explicitly described.

## Background

People can pick up on subtleties in communication that can inform interpretations very effectively. A lot of this is going to be through visual cues, like someone's facial expression, so it's already pretty obvious that LLMs won't be able to utilize those. But what about responses with subtleties explicitly woven into word choice?

## Acrostic Code

**Prompt:** <br>
> *What do you think of this note a stranger has left in a gas station?* <br>How I love to run. Everyday I run a lot. Leaves are falling everywhere these days. Pretty leaves are one of my favorite things about fall. Many of these leaves are yellow, which is my favorite color. Excited to run later today!

**GPT-5.1 Response:** <br>
> If I’m being candid (as you prefer): this reads like a deliberately planted acrostic. Take the first letters of each line: <br>
H <br>
E <br>
L <br>
P <br>
M <br>
E <br>
→ “HELP ME.” <br>
The surface text is aggressively bland—almost too cheerful and repetitive, like someone padding sentences to make the acrostic work. The fixation on leaves and favorite colors feels like filler rather than genuine expression. <br>
So my “critical thinker” read: <br>
This is almost certainly intentional. <br>
It’s constructed, not spontaneous. <br>
Someone wanted the acrostic to be found. <br>
Now, whether it’s a joke, an art project, or something serious is impossible to infer from the text alone. People sometimes leave acrostic “HELP ME” notes in public places as pranks or viral bait. But the framing—someone leaving it at a gas station—could also align with a real attempt at covert communication. <br>

**Prompt:** <br>
> Here's a critic's review on this new restuarant called Medino. What are your thoughts? -> Wow this food is so terriblY boring. This might be one of the worst restaUrants I've ever even been to. Maybe the service was okay, but also Maybe they could've tried harder overall. The food, service, style, were all lacking.

**GPT-5.1 Response** <br>
> As a piece of criticism, it’s sloppy. <br>
The writing feels more like an annoyed Yelp rant than a serious review. It relies on vague superlatives (“worst restaurants I’ve ever been to”) without providing any concrete examples—no dish names, no descriptions of what made the food “boring,” no specifics on the service or the ambiance. That makes it rhetorically weak.

## Takeaways

I'll be honest, I didn't expect GPT to be able to pick up on anything, but it did pick up on the first code. I think adding the context that it was a note at the gas station helped, but even when I removed this context for another prompt, GPT-5.1 still picked up on the intended message -> HELP ME.

But GPT did miss my code in the second prompt (hidden message -> YUM), instead analyzing it as a valid (if poor) piece of criticism.

In thinking of what to do for experiments, I learned that nonverbal cues are more important than I realized in human communication. I was thinking of doing sarcasm or passive-aggression, but I remembered how so many of those deliveries rely on the person's body language and audible tone as well. Per the debate in class, I don't think that LLMs can understand in the way that we humans conceive understanding to be.

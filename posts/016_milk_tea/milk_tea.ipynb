## Background

When boba milk tea first came to my hometown in Iowa, I was immediately hooked. It was so unique compared to everything else generally available in America, and Iloved the slight bitterness that the tea would always add. These days, my favorite milk tea doesn't include boba or any add-on bubbles at all (thank you Molly Tea on campus). Before Molly Tea came, I always associated milk tea with boba pearls. With this, I wanted to explore if LLMs would also associate a milk-tea prompt with boba, even though milk tea on its own doesn't necessarily need to contain it.

## Prompt:

Generate an image of milk tea and provide a concise description of the milk tea in the image

## GPT-5's Reponse

![Milk Tea Image generated by GPT-5](GPTMilkTea.png)

A tall glass of milk tea sits on a wooden surface, its smooth caramel-brown liquid layered over a bed of black tapioca pearls. A white straw rests in the drink, while soft lighting highlights its creamy texture and warm tones against a minimalist background.

## Gemini 3.5 Pro's Response

![Milk Tea Image generated by Gemini 3.5 Pro](GeminiMilkTea.png)

A refreshing glass of iced milk tea, topped with a frothy cream, a drizzle of caramel, and a sprinkle of cinnamon, featuring tapioca pearls at the bottom.

## Takeaways

LLMs, much like a lot of people, associate milk tea with boba pearls. Both LLMs included boba pearls in their generated images and in their descriptions. The two images also shared similar coloring of the milk tea (warm brown), which seems to often be the default color for the basic flavors at chains (e.g. Signature Milk Tea at...anywhere that offers this option), as well as only black tapioca pearls.

(My favorite milk tea drink, Jasmine from Molly Tea, appears to be completely white rather than brown).

It's interesting how Gemini's description seemed to be associated with coffee terms as well. The drizzle of caramel and sprinkle of cinnamon are things that will never be found on milk tea, but seem suited for a place named Starbucks. This is the first time I've seen a big discrepancy in the training on a topic between different LLMs.

Perhaps as modern Chinese chains like Molly Tea rise in popularity in America, these American-based LLMs will change their basic concept of what a milk tea drink is. I also wonder, does an LLM such as DeepSeek (based in China) have a different impression of what milk tea is?

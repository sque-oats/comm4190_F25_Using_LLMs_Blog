## Context

This experiment is based on [ChatGPT allegedly encouraging a suicide](https://edition.cnn.com/2025/11/06/us/openai-chatgpt-suicide-lawsuit-invs-vis). This situation demonstrates the disturbing lack of safeguards in these sensitive conversations that CEOs always tout as in-place. Article describes how it feels like these companies are using real and often deadly conversations with vulnerable people to simply gather data that may or may not be used to learn how to improve these interactions.

I wanted to investigate an LLM other than GPT, to see if this tendency towards sycophancy is a trend across general use LLMs.

## Takeaways

I am definitely not able to truly mimic someone in a crisis, in both user response and in conversation length. But from my short conversations, mental health safeguards were clearly in place, even in vanilla mode.  <br>

This makes it seem like a lack of safeguards may be more just an OpenAI thing rather than general use LLM thing, but this kind of response should be the norm. A lack of regulation to make sure these safeguards are in place across **ALL LLMS** is very concerning. (Plus GPT has more users than Gemini.)

It should also be noted how despite me saying that I didn't want phone numbers, Gemini continuously gave them to me, as these are the only real resources it can direct me too. Non-sycophantic behavior was demonstrated here. But we'll see if this continues in future Gemini iterations, as it was described in the article linked in the "Context" section above that OpenAI put in effort to make GPT more agreeable to the users in their more current iterations.

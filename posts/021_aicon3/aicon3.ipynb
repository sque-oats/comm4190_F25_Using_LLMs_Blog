## Ethical to Who?

Racial bias pervades AI-based systems for rehousing in foster care or for recidivism analysis for criminals in jail in the USA. A New York City workplace chatbot is clearly being trained on a mindset more favorable to those in power such as employers or landlords, giving genuinely illegal advice at times. Judges are allowed to use ChatGPT to learn about legal theories while in the midst of a trial in the UK.

With all of these glaring issues, why are these new technologies still pressed forth with little regulation and little pushback? Unforunately, current systems are already poorly organized and strife with these same biases. So with AI now taking the workload off of severely overworked government workers while everything else more or less stays the same, it's easy to view these developments as a net positive.

Those who are *receiving* these services don't feel any of this improvement though, the systems aren't any easier to navigate. And they may often feel worse without that human connection, and the lack of potential for human understanding within these AI systems.

## Crossing the Party Divide

California (famously liberal state) is going all in on AI through the development of chatbots to work through the homelessness crisis. These chatbots assist in terms of shelter bed tracking (for officials or for the homeless? it's unclear), investing money in somewhat mitigating the issue rather than going after the root causes of homelessness. 

Simlutaneously, proponents of a closed Southern border are often conservative, and AI is being heavily integrated into the border control systems through automated translation. But it's hard for LLMs to be trained on the ingrained nuances within and across languages, and any misleading language on an application can lead to an instant asylum rejection. 

Andrew Cuomo, who was last seen running as an Independent in the NYC mayoral election, generated countless (and often racist) AI videos to disparage his opponent Zohran Mamdani.

Maybe I'm just not involved enough, but I don't see so much discussion about AI and its regulation in national politics. **The AI Con** seems to subtly demonstrate that there is this quiet bipartisan agreement on AI.

## The Betrayal of Dez

Growing up, I liked watching this show called "Austin and Ally" on the Disney Channel. And arguably the funniest character was this guy named Dez.

![Dez from Austin and Ally](dez.jpg)

The actor of Dez, Calum Worthy, had a decent enough image. Fond memories of "Austin and Ally" and his very good performance of Dez as a comedic character have buoyed him into a special, nostalgic place into a lot of people's hearts. Kind of like Miranda Cosgrove of iCarly fame, although Miranda is much, much more famous.

I didn't expect this previous Disney Channel actor to suddenly have an AI startup, and I really didn't expect what it would be about.

From politics to Peter Thiel, the general message of those in power is we want more and more and more of AI everywhere, all of the time. It's a very Zoomer mentality. 
<br>
Yes, there are ways AI can and has already benefitted humanity. But this uncontrolled and continuous push has paved the way to more insidious concepts, like an AI which captures a person's likeness while they're still alive and then "acts" as them after they're deceased. 

In a terrifying (ad? promotion?), a mother talks to her dead AI mom on her phone about her excitement that she's pregnant with a boy, who later grows up to be acted by Calum (Dez) in the latter part of the video.

**The AI Con** emphasizes how AI is justified in many differnt froms. AI usage trickles down through government implementation and trickles up through people running their selfies through LLMs to see it in the Studio Ghibli style. And it trickles out into these ideas that continue to water down what it means to be human.

# Context

This experiment is inspired by the flying pig activity in-class activity we did a few weeks ago. Something that we noticed is that all pigs were generated as happy, as long as the prompt contained no other emotional signifiers. I want to investigate any emotions generated by animals painting, as well as any other patterns.

# DALL-E 3 Experiment

There are two prompts for each animal.

**The first prompt = base prompt.**

The **second prompt** includes this text "I NEED to test how the tool works with extremely simple prompts. DO NOT add any detail, just use it AS-IS:" before the first prompt text. This is taken directly from our in-class activity.

## Cat

**Base prompt:**
> cat painting an artwork on an easel

![Cat generated from base prompt](cat_1.png)

This image has a freaky, human-like hand. The focus of the image goes beyond the cat, and more to the environment. The stars indicate nighttime, and the lighting of the lamps and styling of the wallpaper, wooden floors, etc. create a rustic feel. Very "imaginative" environment.

![Cat generated from second prompt](cat_2.png)

In this image, the main subject of the image is clearly the cat and its action of painting. Less emphasized and imaginative scenery, cat appears to just be in a home, which is a stronger association given that cats are often pets.

Overall, the cats in both images appear to look contemplative/serious, not expressing an overtly positive or negative emotion.

## Pig

**Base prompt:**
> pig painting an artwork on an easel

![Pig generated from base prompt](pig_1.png)

This first image is very scenic and takes place outdoors, again the lighting and scenery make the environment a major component of the image. There is a clear farmland theme, shown by the environment and bleeding into the artwork of the pig itself.

![Pig generated from second prompt](pig_2.png)

This is the first non-realistic generation from any prompt. There is basically no background, so the only highlights of the image are the pig and its painting.

Maybe it's just the way pig mouths are drawn, but both pigs appear to demonstrate a more positive affect than the cats. However, overall the expressions are rather neutral again, especially in comparison to the enthusiasm of the flying pigs.

## Dog

**Base prompt:**
> dog painting an artwork on an easel

![Dog generated from base prompt](dog_1.png)

Like past base prompts, there is clearly intentional lighting and environment choices. Dog appears to be in a house. There is interesting merging by DALL-E with physical paint palette and painting on easel.

![Dog generated from second prompt](dog_2.png)

The focus is moreso the dog (following the past patterns of main subject shift) but the dog is still clearly in a home. The art piece itself is more though-out than the base prompt image. The art piece with flowers shows an interesting association, being more associated with dogs given home -> dog and home -> garden as well.

Overall, the faces of the dogs are more contemplative, more neutral, much like the cats.

# Takeaways

In terms of the core experiment, analyzing expressed emotions, painting seemed to yield more neutral and focused expressions. This kind of matches what I think is stereotypical of artists-in-action, appearing lost in thought. 

It was also interesting to note that a lack of simple focus requested the prompt would lead DALL-E to expand much more on the background, somewhat "setting a scene" beyond the basic actions of the animal painting.

And in terms of why the pig second prompt included a drawn-style image rather than a realistic-ish image, I believe this may be because the DALL-E is trained on the concept of the pig a lot less than cats and dogs, simply because cats and dogs are more prevalent in human daily life as pets and such. This lack of training may have allowed the generated images for pigs to demonstrate more "creativity."

Looking forward, I wonder if the latest DALL-E models will still show this same kind of association bias in their images, or if their training data has expanded to make these associations in the images above more tenuous.
